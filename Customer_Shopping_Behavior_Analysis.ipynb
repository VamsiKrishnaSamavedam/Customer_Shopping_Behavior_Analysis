### We have 37 Review rating Null values instead of directly filling them we will group them  according to categories and then fill the nulls in reviews for not getting biased reviews.

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Lets convert the column names to camel case formating for easy understanding of the columns 
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# As age is the continous feature create bins for better analysis for age feature.
labels=['Young-Adult','Adult','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

1. We have a column called Frequency of Purchases where it is showing in text format regarding purchasing lets convert them to numeric format for better understanding.

```python
df['frequency_of_purchases'].value_counts()
```

```python
# create a column called purchace_frequency_days
freqmap={
    'Every 3 Months':90,
    'Annually':365,
    'Quarterly':90,
    'Monthly':30,
    'Bi-Weekly':14,
    'Fortnightly':40,
    'Weekly':7
}
df['purchace_frequency_days']=df['frequency_of_purchases'].map(freqmap)
df.head(5)
```
